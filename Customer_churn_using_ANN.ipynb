```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#remove the redundant columns from the dataset
df.drop('customerID',axis='columns',inplace=True)
```

```python
#droping the rows with null values
df[pd.to_numeric(df.TotalCharges,errors='coerce').isnull()]
```

```python
#formation of new dataframe where Totalcharges is no longer ''
df1 = df[df.TotalCharges!=' ']
df1.shape
```

```python
#since the datatype of df1.TotalCharges is Object we need to consider the numeric Value of it
df1.TotalCharges=pd.to_numeric(df1.TotalCharges)
df1.TotalCharges.dtypes
df1.TotalCharges.values
```

```python
#histogram for showing the customer churn

tenure_churn_yes=df1[df1.Churn=='Yes'].tenure
tenure_churn_no=df1[df1.Churn=='No'].tenure

#plot the graph
plt.xlabel("tenure")
plt.ylabel("No of Customers")
plt.hist([tenure_churn_yes,tenure_churn_no],color=['green','red'],label=['Churn_Yes','Churn_No'])
plt.legend()
plt.show()
```

```python
def print_unique_col_values(df):
       for column in df:
            if df[column].dtypes=='object':
                print(f'{column}: {df[column].unique()}') 
```

```python
print_unique_col_values(df1)      
#Here No internet services is same as no itself
#replace no internet services with No over here
```

```python
df1.replace('No phone service','No',inplace=True)
df1.replace('No internet service','No',inplace=True)
print_unique_col_values(df1)      
```

```python
#Encoding 

to_numeric_cols=['Partner','Dependents','PhoneService','MultipleLines','OnlineSecurity','OnlineBackup',
                  'DeviceProtection','TechSupport','StreamingTV','StreamingMovies','PaperlessBilling','Churn']
for col in to_numeric_cols:
  df1[col].replace({'Yes':1,'No':0},inplace=True)
```

```python
df1.dtypes
```

```python
df1.dtypes
```

```python
df1['gender'].replace({'Female':1,'Male':0},inplace=True)
df1
```

```python
#encoding scheme for categorical datatype
df2 = pd.get_dummies(data=df1, columns=['InternetService','Contract','PaymentMethod'])
```

```python
df2.columns
```

```python

df2.dtypes
```

```python
df1['TotalCharges']
```

```python
 cols_to_scale=['tenure','MonthlyCharges','TotalCharges']

 from sklearn.preprocessing import MinMaxScaler
 scaler=MinMaxScaler()
 
 df2[cols_to_scale]=scaler.fit_transform(df2[cols_to_scale])
```

```python
for col in df2:
  print(f'{col}:{df2[col].unique()}')
```

```python
#training and testing data
X=df2.drop('Churn',axis='columns')
y=df2['Churn']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=5)
```

```python
#shape of training and testing data
print("Training Samples:",X_train.shape[0])
print("Testing Samples:",X_test.shape[0])
```

```python
import tensorflow as tf
from tensorflow import keras

#generation of a keras sequential model
model=keras.Sequential([
    #we have considered 20 hidden layer along with input layer
    keras.layers.Dense(20,input_shape=(26,),activation='relu'),
    #we have one output layer
    keras.layers.Dense(1,activation='sigmoid')
])
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
#fitting of above generated model
model.fit(X_train,y_train,epochs=10)
```

```python
#evaluate the model
model.evaluate(X_test,y_test)
```

```python
#y_predition array
y_p=model.predict(X_test)
#print(y_pred.dtype)

#conversion of 2d to 1d array
y_pred=[]
for element in y_p:
  if(element<0.5):
    y_pred.append(0)
  else:
    y_pred.append(1)
      
```

```python
#generation of classification report

from sklearn.metrics import confusion_matrix,classification_report
print(classification_report(y_test,y_pred))
```

```python
#generation of confusion matrix
import seaborn as sns
cm=tf.math.confusion_matrix(labels=y_test,predictions=y_pred)
plt.figure(figsize=(10,7))
sns.heatmap(cm,annot=True,fmt='d')
plt.xlabel('Predicted Values')
plt.ylabel('Truth Values')
plt.show()
```
